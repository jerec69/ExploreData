```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('physics_particles.csv')
pd.set_option('Display.max_columns', 13)
pd.set_option('Display.max_rows', 50)
df
```

```python
# Convert numeric columns to float
num_cols = ["mass", "mass_lower", "mass_upper", "width", "width_lower", "width_upper"]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")

#double check
print(df.dtypes)
```

```python
# Mass of Particles
plt.figure(figsize=(12,6))
ax = sns.barplot(x="name", y="mass", data=df)

plt.xticks(rotation=90)
plt.title("Mass Distribution of the Particles")
plt.xlabel("Particle Name")
plt.ylabel("Mass (MeV/c^2)")

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f"{height:,.0f}", 
                    (p.get_x() + p.get_width()/2., height),
                    ha="center", va="bottom", fontsize=7)

plt.show()
```

The bar chart above shows the mass distribution of elementary and composite particles.  
- We can clearly see that the top quark has the highest mass (~172,500 MeV/c^2), while the electron and neutrinos have extremely small masses.  
- The large difference in scale highlights why a logarithmic scale is often used in particle physics.  
- This kind of visualization helps compare the relative magnitudes of particle masses, making patterns more interpretable for both physicists and data analysts.

```python
#Average of Quark Composition
Average_mass = df.groupby('quarks')['mass'].mean().sort_values()
sns.barplot(x = Average_mass.index, y = Average_mass.values)
plt.xticks(rotation = 90)
plt.title('Average Mass Per Quark Composition')
plt.show()
```

The bar chart illustrates how the average particle mass depends on quark composition. Particles made of only light quarks (u and d), such as protons (uud) and neutrons (udd), have relatively low and stable masses. In contrast, the inclusion of heavier quarks like charm (c), bottom (b), and especially top (t) leads to a dramatic rise in the average particle mass.

```python
#Charge vs Mass
sns.boxplot(x='charge', y='mass', data=df)
plt.yscale("log")
plt.title("Mass by Particle Charge")
plt.xticks(rotation = 90)
plt.show()
```

The boxplot shows the distribution of particle masses grouped by their electric charge, plotted on a logarithmic scale. We can observe that:

- Neutral particles (charge = 0) span the widest range of masses, from nearly massless neutrinos up to very heavy quarks.

- Charged particles (+1 or 1) tend to have a more limited mass range, clustering in the light to medium spectrum.

The log scale reveals clear differences between groups, highlighting that charge alone does not directly determine mass, but neutral particles are more diverse in terms of possible mass values.

This suggests that particle charge is not strongly correlated with mass, but neutrality allows for the existence of both extremely light (neutrinos, photons) and extremely heavy (top quark) states.

```python
# Mass vs Width with Error Bars

plt.errorbar(df['mass'], df['width'],
             xerr=[df['mass_lower'], df['mass_upper']],
             yerr=[df['width_lower'], df['width_upper']],
             fmt='o', alpha=0.7)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Mass (MeV/c^2)")
plt.ylabel("Width (MeV)")
plt.title("Mass vs Width with Uncertainties")
plt.show()
```

The scatter plot with error bars shows the relationship between particle mass and decay width, both on logarithmic scales. The error bars capture the uncertainties in experimental measurements.

- We can observe a general positive trend: heavier particles tend to have larger decay widths, meaning they are less stable and decay more quickly.

- Light particles (like electrons, photons, and neutrinos) cluster at the lower left, with very small or zero widths, reflecting their stability.

- The large uncertainties for some heavy particles (e.g., top quark) highlight the experimental challenges in measuring very short-lived states.

This visualization reinforces the principle that massive particles are typically unstable, and connects particle physics data with the concept of lifetime vs. mass hierarchy in the Standard Model.

<h3>Insight and Conclusion:</h3>

From the different visualizations and analyses, we can summarize the following key points:

1. Mass Distribution

    - The particle mass spectrum spans an extreme range  from nearly massless neutrinos and photons, up to the very massive top quark (~172,500 MeV/c^2).

    - A logarithmic scale is essential to compare particles meaningfully because of these vast differences.

2. Quark Composition and Mass

    - Particles composed of light quarks (u, d) are stable and low in mass, forming the basis of ordinary matter (e.g., protons, neutrons).

    - Adding heavier quarks (s, c, b, t) drastically increases the particle mass, with the top quark dominating.

    - This highlights the hierarchical mass structure of quarks in the Standard Model.

3. Charge vs Mass

    - Neutral particles exhibit the widest range of masses, from the lightest neutrinos to the heaviest quarks.

    - Charged particles are more concentrated in lighter to mid-mass ranges, suggesting that neutrality allows for both very light and very heavy states.

4. Mass vs Width Relationship

    - A general trend shows that heavier particles have larger decay widths, meaning they are shorter-lived and less stable.

    - Stable particles (like electrons and protons) cluster at low widths, while unstable heavy particles (like top quarks) have high widths and uncertainties.
